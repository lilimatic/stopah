```python
#Split into train-test based on treatment 

train = stopah.loc[stopah['Prednisolone']==0].drop(['Prednisolone'], axis=1)

test= stopah.loc[stopah['Prednisolone']==1].drop(['Prednisolone'], axis=1)

train.reset_index()
test.reset_index()
```

## Training at 28 days

```python
X_train = train[baseline]
y_train = train['D28_DTH']

X_test = test[baseline]
y_test = test['D28_DTH']
```

```python
## GAM
```

```python
gam = LogisticGAM(f(0,lam=0) + f(1,lam=0) + f(2,lam=0) + s(3) + s(4) + s(5) + s(6) + s(7) + s(8) + s(9) + s(10) 
    + s(11) + s(12) + s(13) + s(14) + s(15) + s(16) + s(17) + s(18) + s(19) + s(20) + s(21) 
    + s(22) + s(23) + s(24) + s(25) + s(26) + s(27) + s(28) + s(29) + s(30), max_iter=2000).fit(X_train.values,y_train.values)

#gam
```

```python
gam.summary()
```

```python
gam.terms
```

```python
for i, term in enumerate(gam.terms):
    if term.isintercept:
        continue

    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)

    plt.figure()
    plt.plot(XX[:, term.feature], pdep)
    plt.plot(XX[:, term.feature], confi, c='r', ls='--')
    plt.title(repr(term))
    plt.show()
```

```python
pred_prob = gam.predict_proba(X_test)

auc_score = roc_auc_score(y_test, pred_prob)

print('Training accuracy: ', np.mean(gam.predict(X_train) ==y_train)*100) 
print ('Test accuracy:', np.mean(gam.predict(X_test) ==y_test)*100)
print('AUC Score:',(auc_score)*100)
```

```python
gam = LogisticGAM(s(3) + s(4) + s(5) + s(6) + s(7) + s(8) + s(9) + s(10) 
    + s(11) + s(12) + s(13) + s(14) + s(15) + s(16) + s(17) + s(18) + s(19) + s(20) + s(21) 
    + s(22) + s(23) + s(24) + s(25) + s(26) + s(27) + s(28) + s(29) + s(30), max_iter=2000,lam=4).fit(X_train.values,y_train.values)

#gam
```

```python
pred_prob = gam.predict_proba(X_test)

auc_score = roc_auc_score(y_test, pred_prob)

print('Training accuracy: ', np.mean(gam.predict(X_train) ==y_train)*100) 
print ('Test accuracy:', np.mean(gam.predict(X_test) ==y_test)*100)
print('AUC Score:',(auc_score)*100)
```
