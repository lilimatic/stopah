```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.metrics import roc_auc_score

from sklearn.model_selection import cross_val_score

from sklearn.metrics import classification_report

# Load some other sklearn functions
from sklearn.model_selection import train_test_split


# Import other libraries
import pandas as pd
import numpy as np
```

```python
from data import *
```

```python
#Read data

df, X, y, X_train, X_test, y_train, y_test = dataset('baseline',28,0,'train-test',0.3,'cc')
```

```python
logit_cv = LogisticRegressionCV(cv=10,solver='liblinear')

logit_net = LogisticRegressionCV(cv=10, penalty='elasticnet', l1_ratios=[0.1, 0.5, 0.9], solver='saga')
```

```python
# Train the models
logit_cv.fit(X_train, y_train)
logit_net.fit(X_train, y_train)

```

```python
# Test the models
print("Logistic Regression: {} || Elasticnet: {}".format(logit_cv.score(X_test, y_test), logit_net.score(X_test, y_test)))

# Print out some more metrics
print("Logistic Regression")
print(classification_report(y_test, logit_cv.predict(X_test)))
print("Elastic Net")
print(classification_report(y_test, logit_net.predict(X_test)))
```

```python
pred_prob = logit_cv.predict_proba(X_test)
auc_score = roc_auc_score(y_test, pred_prob[:,1])
```
